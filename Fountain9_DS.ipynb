##  <p style='color:red'>use 'Purchase Order 1' to calculate the following two things:<p/>
### 1) add column 'Lead Time':Lead Time is the difference between the 'Ordered Date' and the 'Received Date' in days
### 2) calculate 'Lead Time' Average and Varience for each Product ID-Branch ID-Supplier ID combination

```python
# first check missing values
po_data[po_data['Received Date'].isna()]
```

```python
# fill missing values
po_data['Received Date'].fillna(method='bfill',inplace = True) # bcz all orders are recived at the same date
po_data
```

```python
# add column 'Lead Time':Lead Time is the difference between the 'Ordered Date' and the 'Received Date' in days
po_data['Lead Time'] = pd.to_datetime(po_data['Received Date'] )-pd.to_datetime(po_data['Ordered Date'])
po_data['Lead Time']
```

```python
# here datatype of po_data['Lead Time'] is timedelta64. we have to convert it into int
po_data['Lead Time']=(po_data['Lead Time']/np.timedelta64(1,'D')).astype(int)
po_data['Lead Time']
```

```python
po_data
```

### 2) calculate 'Lead Time' Average and 'Lead Time' Varience for each Product ID-Branch ID-Supplier ID combination


```python
po_data.groupby(['Product ID','Branch ID','Supplier ID'])['Lead Time'].agg([np.mean,np.var])
```

# <p style = 'color:red'>Use 'Select' sheet and add new column'Result' to it.</p>
### a) For all rows where the category is 'Biscuits' populate the 'Result' column with the value of 'Quantity' column.
### b) for all other column populate the 'Result' column with the value of 'Price' column.

```python
# show data of 'Select' column
s_data
```

```python
# a) For all rows where the category is 'Biscuits' populate the 'Result' column with the value of 'Quantity' column.
s_data['Result']=s_data[s_data['Product Category'] == 'Biscuits']['Quantity']
s_data
```

```python
#  b) for all other column populate the 'Result' column with the value of 'Price' column.
s_data.fillna({'Result':s_data['Price']},inplace=True)
s_data
```
